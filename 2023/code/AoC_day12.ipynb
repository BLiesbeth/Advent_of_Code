```python
@cache
def count_matches(pattern, size, numbers):
    #print("input: ", pattern, size, numbers)
    
    if len(numbers) == 0:
        if all(c in '.?' for c in pattern):
            return 1
        return 0

    num = numbers[0]
    rest = numbers[1:]
    after = sum(rest) + len(rest)
    
    count = 0

    for before in range(size-after-num+1):
        cand = '.' * before + '#' * num + '.'
        #print(cand)
        if all(c0 == c1 or c0=='?' for c0,c1 in zip(pattern, cand)):
            count += count_matches(pattern[len(cand):], size-num-before-1, rest)

    return count
```

```python
# Part 1

res = 0
for line in puzzle:
    fields, numbers = line.split()
    fields = fields + '.'
    numbers = [int(x) for x in numbers.split(',')]
    res += count_matches(fields, len(fields), tuple(numbers))
print(res)
```

```python
# Part 2

res = 0
for line in puzzle:
    fields, numbers = line.split()
    fields = '?'.join([fields] * 5) + '.'
    numbers = [int(x) for x in numbers.split(',')] * 5
    res += count_matches(fields, len(fields), tuple(numbers))
print(res)
```
