```python
import pandas as pd
import numpy as np
import os
from aocd import get_data

import sys
```

```python
puzzle = get_data(day=11, year=2023)

file = "/Users/liesbeth_bogaert/Documents/AoC/2023/input/input_day11.txt"
lines = open(file, "w").write(puzzle)

puzzle = puzzle.splitlines()
print(puzzle)
```

```python
grid = [[c for c in line] for line in puzzle]

def is_row_empty(i):
    return all([grid[i][j] == "." for j in range(len(grid[i]))])

def is_col_empty(j):
    return all([grid[i][j] == "." for i in range(len(grid))])

empty_rows = [i for i in range(len(grid)) if is_row_empty(i)]
empty_cols = [j for j in range(len(grid[0])) if is_col_empty(j)]

print(empty_rows)
print(empty_cols)
```

```python
locations = [(i, j) for i in range(len(grid)) for j in range(len(grid[i])) if grid[i][j] == "#"]
pairs = [(g1, g2) for g1 in locations for g2 in locations if g1 < g2]
```

```python
def calc_distances(g1, g2, empty):
    g1_x, g1_y = g1
    g2_x, g2_y = g2
    horizontal = abs(g1_y - g2_y)
    vertical = abs(g1_x - g2_x)
    
    for i in range(min(g1_x, g2_x) + 1, max(g1_x, g2_x)):
        if i in empty_rows:
            vertical += empty - 1
            
    for j in range(min(g1_y, g2_y) + 1, max(g1_y, g2_y)):
        if j in empty_cols:
            horizontal += empty - 1

    return horizontal + vertical
```

```python
distances = [calc_distances(g1, g2, 2) for g1, g2 in pairs]

print(sum(distances))
```
