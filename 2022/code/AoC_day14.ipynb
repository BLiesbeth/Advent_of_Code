```python
lines = open(file).read().strip().splitlines()
#print(lines)
```

```python
grid = {}
floor = 0

for line in lines:
    stones = line.split(' -> ')
    stones = [tuple(map(int, x.split(","))) for x in stones]

    x, y = stones[0]
    grid[x,y] = u"\u2593"
    floor = max(floor, y)
    
    for nx, ny in stones:
        while x != nx or y != ny:
            dx = (nx > x) - (x > nx)
            dy = (ny > y) - (y > ny)
            x = x + dx
            y = y + dy
            grid[x,y] = u"\u2593"
            floor = max(floor, y)
print(floor)
```

```python
time = 0

overflow = True

while overflow:
    x, y = 500, 0
    if grid.get((x, y)) == u"\u2592":
        break
    
    while (x,y) not in grid:
        if y > floor:
            overflow = False
            break
        elif (x, y+1) not in grid: #down
            y += 1
        elif (x-1, y+1) not in grid: #left
            x -= 1
            y += 1
        elif (x+1, y+1) not in grid: #right
            x += 1
            y += 1
        else:
            grid[x,y] = u"\u2592"
            time += 1
    
print(time)
```
